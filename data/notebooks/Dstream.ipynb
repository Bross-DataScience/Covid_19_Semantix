### Exercicios WordCount

```python
 # !hdfs dfs -mkdir /user/bross/stream
```

```python
ssc = StreamingContext(sc, 10)
dstream = ssc.socketTextStream('localhost', 9998)
```

```python
wordcount = dstream.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda key1, key2: key1+key2)
wordcount.pprint()
```
